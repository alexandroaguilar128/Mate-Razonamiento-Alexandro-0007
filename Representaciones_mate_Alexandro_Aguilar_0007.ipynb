#### Problema 1
¡Claro! Vamos a desglosar estos problemas de pensamiento matemático. Resolver ecuaciones es como resolver un misterio: tenemos pistas y debemos encontrar el valor oculto (la variable).Paso 1: Análisis y resolución de las ecuaciones Problema 1: El número misteriosoEnunciado: La suma de un número más 9 es igual a 21.Planteamiento: Definimos nuestra incógnita. Sea $x$ el número que buscamos.Traducción al lenguaje algebraico: La frase "la suma de un número más 9" se escribe como $x + 9$, y "es igual a 21" completa la ecuación:$$x + 9 = 21$$Despeje de la variable: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad realizando la operación contraria (resta):$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar el número misterioso
# resultado_total = 21
# numero_conocido = 9

resultado_total = int(input("Ingrese el resultado: "))
numero_conocido = int(input("Ingrese el número conocido: "))

# Despejamos x: x = resultado_total - numero_conocido
x = resultado_total - numero_conocido

print(f"El número buscado en el Problema 1 es: {x}")
```

#### Problema 2
Problema 2: Las edades de Pedro y AnaEnunciado: Pedro tiene el doble de edad que Ana. Entre los dos suman 36 años.Planteamiento: * Edad de Ana: $a$ Edad de Pedro: $2a$ (porque es el doble)Traducción al lenguaje algebraico: La suma de ambas edades es 36:$$a + 2a = 36$$Despeje de la variable:Primero, agrupamos los términos semejantes: $3a = 36$.El 3 que está multiplicando a la $a$ pasa al otro lado dividiendo:$$a = \frac{36}{3}$$$$a = 12$$Solución: * Ana tiene 12 años.Pedro tiene el doble ($2 \times 12$), es decir, 24 años.Comprobación: $12 + 24 = 36$.

```python
# Problema 2: Edades de Pedro (2a) y Ana (a)
# Ecuación: a + 2a = 36 -> 3a = 36
# suma_edades = 36
suma_edades = int(input("Ingrese la suma de las edades: "))
# Despejamos 'a' (edad de Ana): a = suma_edades / 3
edad_ana = suma_edades / 3
# Calculamos la edad de Pedro
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años.")
print(f"Pedro tiene {int(edad_pedro)} años.")
```
